##### <span style="color:orange">1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.</span>

```python
# Creating a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# append
numbers.append(60)
print("After append:", numbers)

# extend
numbers.extend([70, 80])
print("After extend:", numbers)

# insert
numbers.insert(2, 25)
print("After insert:", numbers)

# remove
numbers.remove(40)
print("After remove:", numbers)

# pop
numbers.pop()
print("After pop:", numbers)

# index
print("Index of 30:", numbers.index(30))

# count
print("Count of 20:", numbers.count(20))

# sort
numbers.sort()
print("After sort:", numbers)

# reverse
numbers.reverse()
print("After reverse:", numbers)

# clear
numbers.clear()
print("After clear:", numbers)
```

# Section C: Exploring Pandas

##### <span style="color:orange">11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then: - Use .info(), .describe() - Select 'Name' and 'City' columns - Drop 'City' column - Fill any missing values in 'Salary' column with the mean - Remove any duplicate rows</span>

##### DataFrame Operations

```python
data = {
    "Name": ["Amit", "Riya", "Karan", "Neha", "Rahul",
             "Amit", "Sneha", "Vikram", "Pooja", "Riya"],
    "Age": [25, 23, 27, 24, 26, 25, 28, 29, 22, 23],
    "City": ["Mumbai", "Pune", "Delhi", "Delhi", "Mumbai",
             "Mumbai", "Pune", "Bangalore", "Chennai", "Pune"],
    "Salary": [50000, 48000, None, 52000, 51000,
               50000, 60000, 65000, None, 48000]
}

df = pd.DataFrame(data)
df
```

##### DataFrame Info

```python
df.info()
```

##### DataFrame Description

```python
df.describe()
```

##### Select Name & City

```python
df[["Name", "City"]]
```

##### Drop City Column

```python
df = df.drop(columns=["City"])
df
```

##### Fill Missing Salary with Mean

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df
```

##### Remove Duplicate Rows

```python
df = df.drop_duplicates()
df
```
