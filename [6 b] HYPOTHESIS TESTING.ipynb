# HYPOTHESIS TESTING

### Background

Bombay Hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners.
The theoretical weekly operating cost is given by:

W = 1000 + 5X


where:

W = weekly operating cost (Rs.)

X = number of units produced per week

Restaurant owners claim that the actual weekly operating costs are higher than what the model predicts.

### Given Data

Sample mean weekly cost (x̄) = 3050

Sample size (n) = 25

Mean units produced (μₓ) = 600

Standard deviation of units produced (σₓ) = 25

Cost per unit = Rs. 5

Significance level (α) = 0.05

### Step 1: State the Hypotheses
This is a right-tailed test because owners claim costs are higher.

Null Hypothesis (H0):
μ = 1000 + 5(600) = 4000

Alternative Hypothesis (H1):
μ > 4000


### Step 2: Calculate the Test Statistic
Population Standard Deviation of Cost
σ = 5 × 25 = 125


Z-test Formula 
Z = (x̄ − μ) / (σ / √n)


```python
import math

# Given values
x_bar = 3050
mu = 1000 + 5*600
sigma = 5 * 25
n = 25

# Z-test statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))
z
```

Z = -38


### Step 3: Critical Value & Significance Level

Significance level (α) = 0.05

Right-tailed Z critical value = 1.645

### Step 4: Decision Rule
Reject H0 if Z > 1.645


#### Comparison
Observed Z = -38
Critical Z = 1.645


#### Decision
Since -38 < 1.645, we FAIL TO REJECT the null hypothesis.


### Step 5: Conclusion
At the 5% significance level, there is no statistical evidence to support the restaurant owners’ claim that the weekly operating costs are higher than predicted by the cost model.

In fact:

The observed mean weekly cost is significantly lower than the theoretical mean.

The cost model does not underestimate operating costs based on the sample data.
